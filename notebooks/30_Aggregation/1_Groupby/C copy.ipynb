# Groupby for bar charts

## Load data

```python
path = '../../../data/EIA/energy_generation_2024_preprocessed.parquet'
```

```python
import pandas as pd
df = pd.read_parquet(path)
df
```

```python
df_base = df.copy()
```

## Calculate temporal properties

```python
df = df_base.copy()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['hour'] = df.index.hour
df['weekday'] = df.index.weekday
df['weekend'] = df.index.dayofweek > 4
```

```python
df_base = df.copy()
```

## Steps

```python
df = df_base.copy()
```

### Aggregate data with groupby

```python
cols_cat = ['hour', 'technology']
cols_num = 'energy'

r = df.groupby(cols_cat)[cols_num].agg(func='sum')
r
```

### Reset index to transform result into a data frame

```python
r = r.reset_index()
r
```
