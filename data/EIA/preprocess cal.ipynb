```python
path = 'demand_cal_historical.parquet'
```

```python
import pandas as pd

df = pd.read_parquet(path)
df
```

```python
df[['period', 'value']].to_csv('demand_cal_historical.csv', index=False)
```

```python
df.filter(like='respondent').value_counts().reset_index().style
```

```python
df[['type', 'type-name']].value_counts()
```

```python
dff = df.query('type == "D"')
```

```python
dff['value-units'].value_counts()
```

```python
df = dff.pivot(index='period', columns='respondent', values='value')
```

```python
mask = df.index.str.startswith('2024-12-')
```

```python
df = df[mask].reset_index()
df
```

```python
df.period = pd.to_datetime(df.period)
```

```python
df.columns = df.columns.set_names(None)
```

```python
df = df.set_index('period')
```

```python
df.index = df.index.strftime('%Y-%m-%dT%H')
```
