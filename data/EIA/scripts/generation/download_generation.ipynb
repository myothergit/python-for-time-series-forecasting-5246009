```python
URL_BASE = "https://api.eia.gov/v2/"
API_KEY = "JDOkVaSFgBpZf05nnVL0cxGvOcBzqRZjDj9JBvZj"
```

```python
endpoint = 'electricity/rto/fuel-type-data/data/'
url = URL_BASE + endpoint + 'data'
```

```python
import requests
import pandas as pd
```

```python
parameters = {
    'api_key': API_KEY,
    'offset': 0,
    'data[]': 'value',
    'frequency': 'local-hourly',
    'sort[0][column]': 'period',
    'sort[0][direction]': 'desc',
    # 'start': '2024-01-01T00',
    # 'end': '2024-12-31T23',
    'facets[respondent][]': ['US48', 'CAL'],
}

data_all = []

while True:
    res = requests.get(url, params=parameters)
    data = res.json()
    data_all.extend(data['response']['data'])
    
    if len(data['response']['data']) < 5000:
        break

    parameters['offset'] += 5000

df = pd.DataFrame(data_all)
```
