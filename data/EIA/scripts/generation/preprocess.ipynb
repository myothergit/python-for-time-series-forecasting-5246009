```python
path = '../../../../data/EIA/energy_generation_2024.parquet'
```

```python
import pandas as pd
df = pd.read_parquet(path).query('respondent == "CAL"')
df
```

```python
df['period'] = pd.to_datetime(df['period'], format='%Y-%m-%dT%H')
df['period'] = df['period'].dt.tz_localize('UTC').dt.tz_convert('America/Los_Angeles')
df = df[['period', 'type-name', 'value']].copy()
df.columns = ['period', 'technology', 'energy']
df.energy = df.energy.astype(float)
df.set_index('period', inplace=True)

df
```

```python
df.to_parquet(path.replace('.parquet', '_preprocessed.parquet'))
```
