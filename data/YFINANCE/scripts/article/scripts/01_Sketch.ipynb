## Row summary

```python
import pandas as pd
```

```python
tma = dfr.mean(axis=1).to_frame(name='Annualized').mul(100)
tmc = dfr.add(1).cumprod(axis=1).sub(1).mul(100).iloc[:,[-1]]
tmc.columns = ['Cumulative']

tm = pd.concat([tma, tmc], axis=1)
```

```python
dfr.columns = dfr.columns.strftime('%Y')
```

```python
df = pd.concat({
    'YEARLY': dfr,
    'TOTAL': tm
}, axis=1)

df
```

## Gradient

```python
from mplcustom import preprocessing
```

```python
gmap = dfr.apply(preprocessing.scale_numbers, axis=1, result_type='expand').values
```

```python
dfs = (dfr
 .mul(100)
 .style
    .format("{:.2f}")
    .background_gradient(cmap='RdYlGn', axis=None, gmap=gmap)
    .format_index(lambda x: x.strftime('%Y'), axis=1)
)
```
